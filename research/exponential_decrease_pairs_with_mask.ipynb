```python
import matplotlib.pylab as plt
%matplotlib inline
```

```python
pairs=[435,300,150,100]

norm_pairs=[]
for i in range(len(pairs)):
    norm_pairs.append(float(pairs[i])/435.0)
    

plt.plot(norm_pairs,'o',c='k')
plt.plot(norm_pairs,c='k')

choices=[94395,44850,11175,4950]

norm_choices=[]
for i in range(len(choices)):
    norm_choices.append(float(choices[i])/94395.0)
    

plt.plot(norm_choices,'o',c='b')
plt.plot(norm_choices,c='b')
```
