```python
"""
3 gaussians to demonstrate peaks in a DEER distance distribution
For CSE 2018 presentation
""""
```

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

font = {'family':'Times New Roman'}
plt.rc('font', **font)
plt.rcParams.update({'font.size':12})
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('savefig', dpi=900)
plt.rc('axes',labelsize=12,labelweight='bold')
plt.rc('legend',fontsize=12)
from pylab import rcParams
```

```python
mu1=2
sigma1=1
s1 = np.random.normal(mu1, sigma1, 100000)
count1, bins1, ignored1 = plt.hist(s1, 100, normed=True)
plt.plot(bins1, 1/(sigma1 * np.sqrt(2 * np.pi)) * np.exp( - (bins1 - mu1)**2 / (2 * sigma1**2) ), linewidth=2, color='r')
```

```python
mu2=1
sigma2=0.5
s2 = np.random.normal(mu2, sigma2, 100000)
count2, bins2, ignored2 = plt.hist(s2, 100, normed=True)
plt.plot(bins2, 1/(sigma2 * np.sqrt(2 * np.pi)) * np.exp( - (bins2 - mu2)**2 / (2 * sigma2**2) ), linewidth=2, color='r')
```

```python
mu3=0
sigma3=0.2
s3 = np.random.normal(mu3, sigma3, 100000)
count3, bins3, ignored3 = plt.hist(s3, 100, normed=True)
plt.plot(bins3, 1/(sigma3 * np.sqrt(2 * np.pi)) * np.exp( - (bins3 - mu3)**2 / (2 * sigma3**2) ), linewidth=2, color='r')
```

```python
plt.fill_between(bins1,0, 1/(sigma1 * np.sqrt(2 * np.pi)) * np.exp( - (bins1 - mu1)**2 / (2 * sigma1**2) ), linewidth=2, color='gray')
plt.fill_between(bins2, 0, 1/(sigma2 * np.sqrt(2 * np.pi)) * np.exp( - (bins2 - mu2)**2 / (2 * sigma2**2) ), linewidth=2, color='gray')
plt.fill_between(bins3, 0, 1/(sigma3 * np.sqrt(2 * np.pi)) * np.exp( - (bins3 - mu3)**2 / (2 * sigma3**2) ), linewidth=2, color='gray')
plt.ylim(0,4)
plt.xlim(-2,5)
```

```python
plt.fill_between(bins2, 0, 1/(sigma2 * np.sqrt(2 * np.pi)) * np.exp( - (bins2 - mu2)**2 / (2 * sigma2**2) ), linewidth=2, color='k',alpha=0.6)
plt.plot(bins2, 1/(sigma2 * np.sqrt(2 * np.pi)) * np.exp( - (bins2 - mu2)**2 / (2 * sigma2**2) ), linewidth=1, color='k')
plt.fill_between(bins3, 0, 1/(sigma3 * np.sqrt(2 * np.pi)) * np.exp( - (bins3 - mu3)**2 / (2 * sigma3**2) ), linewidth=2, color='indigo',alpha=0.6)
plt.plot(bins3, 1/(sigma3 * np.sqrt(2 * np.pi)) * np.exp( - (bins3 - mu3)**2 / (2 * sigma3**2) ), linewidth=1, color='indigo')
plt.fill_between(bins1,0, 1/(sigma1 * np.sqrt(2 * np.pi)) * np.exp( - (bins1 - mu1)**2 / (2 * sigma1**2) ), linewidth=2, color='thistle',alpha=0.6)
plt.plot(bins1, 1/(sigma1 * np.sqrt(2 * np.pi)) * np.exp( - (bins1 - mu1)**2 / (2 * sigma1**2) ), linewidth=1, color='thistle')

plt.fill_between(bins1, -1, 0,color='white')
plt.axhline(0,-2,5,color='k',linewidth=1)

plt.ylim(-1,4)
plt.xlim(-2,5)
plt.yticks([])
plt.tick_params(top='off')
```
