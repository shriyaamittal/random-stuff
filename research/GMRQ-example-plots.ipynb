```python
import matplotlib.pylab as plt
%matplotlib inline

font = {'family':'Helvetica'}
plt.rc('font', **font)
plt.rcParams.update({'font.size':12})
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('savefig', dpi=900)
plt.rc('axes',labelsize=12,labelweight='bold')
plt.rc('legend',fontsize=12)
from pylab import rcParams
```

```python
import random
import numpy as np
```

```python
x1=[50,100,150,200,250,300]
y1=random.sample(np.arange(1.5,4.5,0.02),len(x1))
```

```python
y1
```

```python
x2=[100,150,200,250]
y2=random.sample(np.arange(1.5,4.5,0.02),len(x2))
```

```python
y2
```

```python
plt.plot(x1,y1,color='b',linewidth=2)
plt.plot(x1,y1,'o',color='b',alpha=0.75)

plt.plot(x2,y2,color='r',linewidth=2)
plt.plot(x2,y2,'o',color='r',alpha=0.75)

plt.xlim(40,310)
plt.ylim(1,5)
```

```python
x3=[50,100,150,200,250,300]
y3=[1.98,4.53,4.17,2.68,3.8,2.9]

plt.plot(x3,y3,color='g',linewidth=2)
plt.plot(x3,y3,'o',color='g',alpha=0.75)

plt.xlim(40,310)
plt.ylim(1,5)
plt.tick_params(top="off")
```
