```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
import numpy as np
x=np.arange(20,120,20)
y=np.random.randint(315,355,size=5)
z=np.random.randint(315,355,size=5)
```

```python
font = {'family':'sans-serif'}
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rc('xtick', labelsize=9)
plt.rc('ytick', labelsize=9)
plt.rc('savefig', dpi=300)
plt.rc('axes',labelsize=9)
plt.rc('legend',fontsize=8.1)

plt.rc('figure',figsize=(4,4))

plt.rcParams['axes.linewidth'] = 0.25 

plt.rcParams['xtick.major.size'] = 2
plt.rcParams['xtick.major.width'] = 0.25
plt.rcParams['xtick.minor.size'] = 0
plt.rcParams['xtick.minor.width'] = 0

plt.rcParams['ytick.major.size'] = 2
plt.rcParams['ytick.major.width'] = 0.25
plt.rcParams['ytick.minor.size'] = 0
plt.rcParams['ytick.minor.width'] = 0
```

```python
plt.plot(x,y,label='Test data y')
plt.plot(x,z,label='Test data z')
plt.xticks([20,40,60,80,100])
plt.yticks([315,320,325,330,335,340,345,350,355])
plt.xlabel('Temperature (C)')
plt.ylabel('Peak wavelength (nm)')

plt.legend(frameon=False,loc='lower left',title='T_M (C)')
```
